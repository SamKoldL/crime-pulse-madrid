```python
# Comprobar duplicados en cada serie temporal

duplicados = df.duplicated(
    subset=["municipio", "tipo de crimen", "año", "trimestre"],
    keep=False
)

print("Número de filas duplicadas:", duplicados.sum())

df[duplicados].sort_values(
    ["municipio", "tipo de crimen", "año", "trimestre"]
).head(20)
```

```python
# Crear un orden temporal continuo
orden_trimestre = {
    "Q1": 1,
    "Q2": 2,
    "Q3": 3,
    "Q4": 4
}

df["num_trimestre"] = df["trimestre"].map(orden_trimestre)

# Crear índice temporal consecutivo:
# 2023 Q1 = 0, 2023 Q2 = 1 ... 2025 Q4 = 11
df["periodo_num"] = (
    (df["año"] - 2023) * 4
    + df["num_trimestre"] - 1
)

# Comprobar número de registros por año
print("Registros por año:")
print(df.groupby("año").size())

print("\nMunicipios por año:")
print(df.groupby("año")["municipio"].nunique())

print("\nRango temporal:")
print(df["periodo_num"].min(), "-", df["periodo_num"].max())
```

```python
municipios_2023 = set(df[df["año"] == 2023]["municipio"].unique())
municipios_2024 = set(df[df["año"] == 2024]["municipio"].unique())
municipios_2025 = set(df[df["año"] == 2025]["municipio"].unique())

print("Nuevos en 2024:")
print(sorted(municipios_2024 - municipios_2023))

print("\nNuevos en 2025:")
print(sorted(municipios_2025 - municipios_2024))

print("\nPresentes en 2023 pero ausentes en 2024:")
print(sorted(municipios_2023 - municipios_2024))

print("\nPresentes en 2024 pero ausentes en 2025:")
print(sorted(municipios_2024 - municipios_2025))
```

```python
# ==========================================
# CREACIÓN DEL DATASET TEMPORAL PARA EL MODELO
# ==========================================

# Creamos una copia para no modificar el dataframe original
df_modelo = df.copy()

# Ordenamos cronológicamente cada serie
df_modelo = df_modelo.sort_values(
    by=["municipio", "tipo de crimen", "periodo_num"]
).reset_index(drop=True)

# Definimos cada serie temporal
grupo_serie = df_modelo.groupby(
    ["municipio", "tipo de crimen"]
)

# Creamos los lags sobre la variable que queremos predecir: conteo
df_modelo["lag_1"] = grupo_serie["conteo"].shift(1)
df_modelo["lag_2"] = grupo_serie["conteo"].shift(2)
df_modelo["lag_3"] = grupo_serie["conteo"].shift(3)
df_modelo["lag_4"] = grupo_serie["conteo"].shift(4)

# Visualizamos el resultado
df_modelo[
    [
        "municipio",
        "tipo de crimen",
        "año",
        "trimestre",
        "conteo",
        "lag_1",
        "lag_2",
        "lag_3",
        "lag_4"
    ]
].head(15)
```

```python
# ==========================================
# FEATURE ENGINEERING TEMPORAL
# ==========================================

# Media de los 4 trimestres anteriores
df_modelo["media_movil_4"] = (
    df_modelo[["lag_1", "lag_2", "lag_3", "lag_4"]]
    .mean(axis=1)
)

# Tendencia reciente:
# cambio entre el último trimestre conocido y el anterior
df_modelo["tendencia_reciente"] = (
    df_modelo["lag_1"] - df_modelo["lag_2"]
)

# Variación respecto al mismo trimestre del año anterior
# Comparamos el trimestre anterior con el dato de hace 4 trimestres
df_modelo["variacion_interanual"] = (
    df_modelo["lag_1"] - df_modelo["lag_4"]
)

# Visualizamos
df_modelo[
    [
        "municipio",
        "tipo de crimen",
        "año",
        "trimestre",
        "conteo",
        "lag_1",
        "lag_2",
        "lag_3",
        "lag_4",
        "media_movil_4",
        "tendencia_reciente",
        "variacion_interanual"
    ]
].head(12)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ==========================================
# BASELINE ESTACIONAL - AÑO 2025
# ==========================================

# Seleccionamos 2025 y solo filas con lag_4 disponible
df_baseline = df_modelo[
    (df_modelo["año"] == 2025) &
    (df_modelo["lag_4"].notna())
].copy()

# Valor real
y_real = df_baseline["conteo"]

# Predicción baseline:
# mismo trimestre del año anterior
y_pred_baseline = df_baseline["lag_4"]

# Métricas
mae_baseline = mean_absolute_error(y_real, y_pred_baseline)
rmse_baseline = np.sqrt(
    mean_squared_error(y_real, y_pred_baseline)
)
r2_baseline = r2_score(y_real, y_pred_baseline)

print("RESULTADOS BASELINE ESTACIONAL 2025")
print("=" * 40)
print(f"MAE:  {mae_baseline:.2f}")
print(f"RMSE: {rmse_baseline:.2f}")
print(f"R²:   {r2_baseline:.4f}")
print()
print("Número de observaciones:", len(df_baseline))
```

```python
# ==========================================
# BASELINE EXCLUYENDO MADRID CAPITAL
# ==========================================

df_sin_madrid = df_baseline[
    df_baseline["municipio"] != "Madrid"
].copy()

y_real_sin_madrid = df_sin_madrid["conteo"]
y_pred_sin_madrid = df_sin_madrid["lag_4"]

mae_sin_madrid = mean_absolute_error(
    y_real_sin_madrid,
    y_pred_sin_madrid
)

rmse_sin_madrid = np.sqrt(
    mean_squared_error(
        y_real_sin_madrid,
        y_pred_sin_madrid
    )
)

r2_sin_madrid = r2_score(
    y_real_sin_madrid,
    y_pred_sin_madrid
)

print("BASELINE 2025 - SIN MADRID CAPITAL")
print("=" * 40)
print(f"MAE:  {mae_sin_madrid:.2f}")
print(f"RMSE: {rmse_sin_madrid:.2f}")
print(f"R²:   {r2_sin_madrid:.4f}")
```

```python
# ==========================================
# WAPE DEL BASELINE
# ==========================================

wape_baseline = (
    np.abs(y_real - y_pred_baseline).sum()
    / np.abs(y_real).sum()
) * 100

wape_sin_madrid = (
    np.abs(y_real_sin_madrid - y_pred_sin_madrid).sum()
    / np.abs(y_real_sin_madrid).sum()
) * 100

print(f"WAPE global:     {wape_baseline:.2f}%")
print(f"WAPE sin Madrid: {wape_sin_madrid:.2f}%")
```

```python
# ==========================================
# MAYORES ERRORES DEL BASELINE
# ==========================================

df_baseline["prediccion_baseline"] = df_baseline["lag_4"]

df_baseline["error_absoluto"] = np.abs(
    df_baseline["conteo"]
    - df_baseline["prediccion_baseline"]
)

df_baseline[
    [
        "municipio",
        "tipo de crimen",
        "año",
        "trimestre",
        "conteo",
        "prediccion_baseline",
        "error_absoluto"
    ]
].sort_values(
    "error_absoluto",
    ascending=False
).head(20)
```

```python
# ==========================================
# PREPARACIÓN DEL DATASET DE MODELADO
# ==========================================

# Nos quedamos solo con filas que tienen 4 lags completos
df_ml = df_modelo.dropna(
    subset=[
        "lag_1",
        "lag_2",
        "lag_3",
        "lag_4",
        "media_movil_4",
        "tendencia_reciente",
        "variacion_interanual",
    ]
).copy()

# Variables predictoras
features = [
    "lag_1",
    "lag_2",
    "lag_3",
    "lag_4",
    "media_movil_4",
    "tendencia_reciente",
    "variacion_interanual",
    "num_trimestre",
    "municipio",
    "tipo de crimen",
]

X = df_ml[features].copy()

# Variable objetivo
y = df_ml["conteo"].copy()

print("Dimensiones X:", X.shape)
print("Dimensiones y:", y.shape)

X.head()
```

```python
# ==========================================
# SPLIT TEMPORAL
# ==========================================

train_mask = df_ml["año"] < 2025
test_mask = df_ml["año"] == 2025

X_train = X[train_mask].copy()
X_test = X[test_mask].copy()

y_train = y[train_mask].copy()
y_test = y[test_mask].copy()

print("Train:", X_train.shape)
print("Test:", X_test.shape)

print("\nAños train:")
print(df_ml.loc[train_mask, "año"].value_counts().sort_index())

print("\nAños test:")
print(df_ml.loc[test_mask, "año"].value_counts().sort_index())
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor

# ==========================================
# PREPROCESAMIENTO
# ==========================================

variables_numericas = [
    "lag_1",
    "lag_2",
    "lag_3",
    "lag_4",
    "media_movil_4",
    "tendencia_reciente",
    "variacion_interanual",
    "num_trimestre"
]

variables_categoricas = [
    "municipio",
    "tipo de crimen"
]

preprocesador = ColumnTransformer(
    transformers=[
        ("num", "passthrough", variables_numericas),
        (
            "cat",
            OneHotEncoder(handle_unknown="ignore"),
            variables_categoricas
        )
    ]
)

# ==========================================
# RANDOM FOREST
# ==========================================

modelo_rf = Pipeline(
    steps=[
        ("preprocesador", preprocesador),
        (
            "modelo",
            RandomForestRegressor(
                n_estimators=300,
                random_state=42,
                n_jobs=-1
            )
        )
    ]
)

# Entrenamiento
modelo_rf.fit(X_train, y_train)

print("Random Forest entrenado correctamente.")
```

```python
# ==========================================
# PREDICCIÓN RANDOM FOREST - 2025
# ==========================================

y_pred_rf = modelo_rf.predict(X_test)

mae_rf = mean_absolute_error(y_test, y_pred_rf)

rmse_rf = np.sqrt(
    mean_squared_error(y_test, y_pred_rf)
)

r2_rf = r2_score(y_test, y_pred_rf)

wape_rf = (
    np.abs(y_test.values - y_pred_rf).sum()
    / np.abs(y_test.values).sum()
) * 100

print("RESULTADOS RANDOM FOREST - 2025")
print("=" * 40)
print(f"MAE:  {mae_rf:.2f}")
print(f"RMSE: {rmse_rf:.2f}")
print(f"R²:   {r2_rf:.4f}")
print(f"WAPE: {wape_rf:.2f}%")
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# ==========================================
# GRADIENT BOOSTING
# ==========================================

modelo_gb = Pipeline(
    steps=[
        ("preprocesador", preprocesador),
        (
            "modelo",
            GradientBoostingRegressor(
                n_estimators=300,
                learning_rate=0.05,
                max_depth=3,
                random_state=42
            )
        )
    ]
)

# Entrenamiento
modelo_gb.fit(X_train, y_train)

print("Gradient Boosting entrenado correctamente.")
```

```python
# ==========================================
# PREDICCIÓN GRADIENT BOOSTING - 2025
# ==========================================

y_pred_gb = modelo_gb.predict(X_test)

mae_gb = mean_absolute_error(
    y_test,
    y_pred_gb
)

rmse_gb = np.sqrt(
    mean_squared_error(
        y_test,
        y_pred_gb
    )
)

r2_gb = r2_score(
    y_test,
    y_pred_gb
)

wape_gb = (
    np.abs(y_test.values - y_pred_gb).sum()
    / np.abs(y_test.values).sum()
) * 100

print("RESULTADOS GRADIENT BOOSTING - 2025")
print("=" * 40)

print(f"MAE:  {mae_gb:.2f}")
print(f"RMSE: {rmse_gb:.2f}")
print(f"R²:   {r2_gb:.4f}")
print(f"WAPE: {wape_gb:.2f}%")
```

```python
# ==========================================
# COMPARACIÓN DE MODELOS
# ==========================================

resultados = pd.DataFrame({
    "Modelo": [
        "Baseline estacional",
        "Random Forest",
        "Gradient Boosting"
    ],
    
    "MAE": [
        mae_baseline,
        mae_rf,
        mae_gb
    ],
    
    "RMSE": [
        rmse_baseline,
        rmse_rf,
        rmse_gb
    ],
    
    "R2": [
        r2_baseline,
        r2_rf,
        r2_gb
    ],
    
    "WAPE (%)": [
        wape_baseline,
        wape_rf,
        wape_gb
    ]
})

resultados.round(3)
```

```python
# ==========================================
# VALIDACIÓN RECURSIVA - PREPARACIÓN
# ==========================================

# Histórico conocido: solo hasta finales de 2024
df_historico = df_modelo[
    df_modelo["año"] <= 2024
].copy()

# Series que queremos predecir en 2025
series_2025 = (
    df_modelo[df_modelo["año"] == 2025]
    [["municipio", "tipo de crimen"]]
    .drop_duplicates()
)

print("Series a predecir:", len(series_2025))
```

```python
# ==========================================
# PREDICCIÓN RECURSIVA 2025
# ==========================================

predicciones_recursivas = []

for trimestre_num in [1, 2, 3, 4]:

    trimestre = f"Q{trimestre_num}"

    filas_prediccion = []

    for _, serie in series_2025.iterrows():

        municipio = serie["municipio"]
        tipo_crimen = serie["tipo de crimen"]

        # Histórico disponible para esta serie
        historial = df_historico[
            (df_historico["municipio"] == municipio) &
            (df_historico["tipo de crimen"] == tipo_crimen)
        ].sort_values(
            ["año", "num_trimestre"]
        )

        # Necesitamos al menos 4 valores anteriores
        if len(historial) < 4:
            continue

        ultimos_4 = historial["conteo"].tail(4).values

        lag_1 = ultimos_4[-1]
        lag_2 = ultimos_4[-2]
        lag_3 = ultimos_4[-3]
        lag_4 = ultimos_4[-4]

        media_movil_4 = np.mean(
            [lag_1, lag_2, lag_3, lag_4]
        )

        tendencia_reciente = lag_1 - lag_2

        variacion_interanual = lag_1 - lag_4

        filas_prediccion.append({
            "municipio": municipio,
            "tipo de crimen": tipo_crimen,
            "lag_1": lag_1,
            "lag_2": lag_2,
            "lag_3": lag_3,
            "lag_4": lag_4,
            "media_movil_4": media_movil_4,
            "tendencia_reciente": tendencia_reciente,
            "variacion_interanual": variacion_interanual,
            "num_trimestre": trimestre_num
        })

    X_futuro = pd.DataFrame(filas_prediccion)

    # Predicción con Gradient Boosting
    predicciones = modelo_gb.predict(X_futuro)

    # Los conteos no pueden ser negativos
    predicciones = np.maximum(
        predicciones,
        0
    )

    X_futuro["conteo_predicho"] = predicciones
    X_futuro["año"] = 2025
    X_futuro["trimestre"] = trimestre

    predicciones_recursivas.append(
        X_futuro.copy()
    )

    # Incorporamos las predicciones al histórico
    # para poder utilizarlas como lags del siguiente trimestre

    nuevas_filas = X_futuro[
        [
            "municipio",
            "tipo de crimen",
            "año",
            "trimestre",
            "num_trimestre",
            "conteo_predicho"
        ]
    ].copy()

    nuevas_filas = nuevas_filas.rename(
        columns={
            "conteo_predicho": "conteo"
        }
    )

    df_historico = pd.concat(
        [
            df_historico,
            nuevas_filas
        ],
        ignore_index=True
    )

# Unimos Q1-Q4
df_pred_recursiva = pd.concat(
    predicciones_recursivas,
    ignore_index=True
)

print(
    "Predicciones generadas:",
    len(df_pred_recursiva)
)

df_pred_recursiva[
    [
        "municipio",
        "tipo de crimen",
        "año",
        "trimestre",
        "conteo_predicho"
    ]
].head(10)
```

```python
# ==========================================
# COMPARAR PREDICCIÓN RECURSIVA CON 2025 REAL
# ==========================================

# Valores reales de 2025
df_real_2025 = df_modelo[
    df_modelo["año"] == 2025
][
    [
        "municipio",
        "tipo de crimen",
        "año",
        "trimestre",
        "conteo"
    ]
].copy()

# Cruzamos predicciones y valores reales
df_validacion_recursiva = df_pred_recursiva.merge(
    df_real_2025,
    on=[
        "municipio",
        "tipo de crimen",
        "año",
        "trimestre"
    ],
    how="inner"
)

print(
    "Observaciones comparadas:",
    len(df_validacion_recursiva)
)

df_validacion_recursiva[
    [
        "municipio",
        "tipo de crimen",
        "trimestre",
        "conteo",
        "conteo_predicho"
    ]
].head(10)
```

```python
# ==========================================
# MÉTRICAS RECURSIVAS POR TRIMESTRE
# ==========================================

resultados_recursivos = []

for trimestre in ["Q1", "Q2", "Q3", "Q4"]:

    datos = df_validacion_recursiva[
        df_validacion_recursiva["trimestre"] == trimestre
    ]

    y_real_trim = datos["conteo"]
    y_pred_trim = datos["conteo_predicho"]

    mae = mean_absolute_error(
        y_real_trim,
        y_pred_trim
    )

    rmse = np.sqrt(
        mean_squared_error(
            y_real_trim,
            y_pred_trim
        )
    )

    r2 = r2_score(
        y_real_trim,
        y_pred_trim
    )

    wape = (
        np.abs(
            y_real_trim.values -
            y_pred_trim.values
        ).sum()
        /
        np.abs(
            y_real_trim.values
        ).sum()
    ) * 100

    resultados_recursivos.append({
        "Trimestre": trimestre,
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2,
        "WAPE (%)": wape
    })

resultados_recursivos = pd.DataFrame(
    resultados_recursivos
)

resultados_recursivos.round(3)
```

```python
# ==========================================
# VALIDACIÓN RECURSIVA SIN MADRID CAPITAL
# ==========================================

df_rec_sin_madrid = df_validacion_recursiva[
    df_validacion_recursiva["municipio"] != "Madrid"
].copy()

y_real_sm = df_rec_sin_madrid["conteo"]
y_pred_sm = df_rec_sin_madrid["conteo_predicho"]

mae_rec_sm = mean_absolute_error(y_real_sm, y_pred_sm)

rmse_rec_sm = np.sqrt(
    mean_squared_error(y_real_sm, y_pred_sm)
)

r2_rec_sm = r2_score(y_real_sm, y_pred_sm)

wape_rec_sm = (
    np.abs(y_real_sm.values - y_pred_sm.values).sum()
    / np.abs(y_real_sm.values).sum()
) * 100

print("VALIDACIÓN RECURSIVA 2025 - SIN MADRID")
print("=" * 45)
print(f"MAE:  {mae_rec_sm:.2f}")
print(f"RMSE: {rmse_rec_sm:.2f}")
print(f"R²:   {r2_rec_sm:.4f}")
print(f"WAPE: {wape_rec_sm:.2f}%")
```

```python
# ==========================================
# RENDIMIENTO POR TIPO DE CRIMEN
# ==========================================

resultados_tipo = []

for tipo, datos in df_validacion_recursiva.groupby("tipo de crimen"):

    y_real_tipo = datos["conteo"]
    y_pred_tipo = datos["conteo_predicho"]

    mae = mean_absolute_error(
        y_real_tipo,
        y_pred_tipo
    )

    rmse = np.sqrt(
        mean_squared_error(
            y_real_tipo,
            y_pred_tipo
        )
    )

    wape = (
        np.abs(
            y_real_tipo.values -
            y_pred_tipo.values
        ).sum()
        /
        np.abs(
            y_real_tipo.values
        ).sum()
    ) * 100

    resultados_tipo.append({
        "Tipo de crimen": tipo,
        "MAE": mae,
        "RMSE": rmse,
        "WAPE (%)": wape,
        "Volumen real": y_real_tipo.sum()
    })

resultados_tipo = pd.DataFrame(
    resultados_tipo
).sort_values(
    "WAPE (%)"
)

resultados_tipo.round(2)
```

```python
# ==========================================
# ANÁLISIS ESPECÍFICO - TRÁFICO DE DROGAS
# ==========================================

df_drogas = df_validacion_recursiva[
    df_validacion_recursiva["tipo de crimen"]
    .str.contains(
        "droga",
        case=False,
        na=False
    )
].copy()

resultado_drogas = (
    df_drogas
    .groupby("trimestre")
    .agg(
        Real=("conteo", "sum"),
        Predicho=("conteo_predicho", "sum")
    )
    .reset_index()
)

resultado_drogas["Error"] = (
    resultado_drogas["Predicho"]
    - resultado_drogas["Real"]
)

resultado_drogas["Error (%)"] = (
    np.abs(resultado_drogas["Error"])
    / resultado_drogas["Real"]
) * 100

resultado_drogas.round(2)
```

```python
# ==========================================
# PREDICCIÓN RECURSIVA Q1-Q3 2026
# ==========================================

# Partimos de todo el histórico REAL disponible hasta Q4 2025
df_historico_2026 = df_modelo[
    df_modelo["año"] <= 2025
].copy()

# Series existentes en 2025 que vamos a proyectar a 2026
series_2026 = (
    df_modelo[df_modelo["año"] == 2025]
    [["municipio", "tipo de crimen"]]
    .drop_duplicates()
)

predicciones_2026 = []

# Predecimos Q1, Q2 y Q3
for trimestre_num in [1, 2, 3]:

    trimestre = f"Q{trimestre_num}"

    filas_prediccion = []

    for _, serie in series_2026.iterrows():

        municipio = serie["municipio"]
        tipo_crimen = serie["tipo de crimen"]

        # Histórico disponible para cada serie
        historial = df_historico_2026[
            (df_historico_2026["municipio"] == municipio) &
            (df_historico_2026["tipo de crimen"] == tipo_crimen)
        ].sort_values(
            ["año", "num_trimestre"]
        )

        # Necesitamos al menos 4 observaciones anteriores
        if len(historial) < 4:
            continue

        ultimos_4 = historial["conteo"].tail(4).values

        lag_1 = ultimos_4[-1]
        lag_2 = ultimos_4[-2]
        lag_3 = ultimos_4[-3]
        lag_4 = ultimos_4[-4]

        media_movil_4 = np.mean(
            [lag_1, lag_2, lag_3, lag_4]
        )

        tendencia_reciente = lag_1 - lag_2

        variacion_interanual = lag_1 - lag_4

        filas_prediccion.append({
            "municipio": municipio,
            "tipo de crimen": tipo_crimen,
            "lag_1": lag_1,
            "lag_2": lag_2,
            "lag_3": lag_3,
            "lag_4": lag_4,
            "media_movil_4": media_movil_4,
            "tendencia_reciente": tendencia_reciente,
            "variacion_interanual": variacion_interanual,
            "num_trimestre": trimestre_num
        })

    # Dataset para realizar la predicción
    X_futuro = pd.DataFrame(filas_prediccion)

    # Predicción con Gradient Boosting
    predicciones = modelo_gb.predict(X_futuro)

    # Evitamos predicciones negativas
    predicciones = np.maximum(
        predicciones,
        0
    )

    X_futuro["conteo_predicho"] = predicciones
    X_futuro["año"] = 2026
    X_futuro["trimestre"] = trimestre

    # Guardamos resultados
    predicciones_2026.append(
        X_futuro.copy()
    )

    # Incorporamos las predicciones al histórico
    # para alimentar el siguiente trimestre
    nuevas_filas = X_futuro[
        [
            "municipio",
            "tipo de crimen",
            "año",
            "trimestre",
            "num_trimestre",
            "conteo_predicho"
        ]
    ].copy()

    nuevas_filas = nuevas_filas.rename(
        columns={
            "conteo_predicho": "conteo"
        }
    )

    df_historico_2026 = pd.concat(
        [
            df_historico_2026,
            nuevas_filas
        ],
        ignore_index=True
    )


# ==========================================
# UNIR RESULTADOS Q1-Q3 2026
# ==========================================

df_pred_2026 = pd.concat(
    predicciones_2026,
    ignore_index=True
)

print(
    "Predicciones 2026 generadas:",
    len(df_pred_2026)
)

print(
    "\nPredicciones por trimestre:"
)

print(
    df_pred_2026["trimestre"].value_counts().sort_index()
)

df_pred_2026[
    [
        "municipio",
        "tipo de crimen",
        "año",
        "trimestre",
        "conteo_predicho"
    ]
].head(20)
```

```python
# ==========================================
# COMPARACIÓN 2025 REAL VS 2026 PREDICHO
# ==========================================

# Datos reales Q1-Q3 2025
real_2025_q1_q3 = (
    df_modelo[
        (df_modelo["año"] == 2025) &
        (df_modelo["num_trimestre"].isin([1, 2, 3]))
    ]
    .groupby("trimestre")["conteo"]
    .sum()
)

# Predicciones Q1-Q3 2026
pred_2026_q1_q3 = (
    df_pred_2026
    .groupby("trimestre")["conteo_predicho"]
    .sum()
)

comparacion_global = pd.DataFrame({
    "Real 2025": real_2025_q1_q3,
    "Predicho 2026": pred_2026_q1_q3
})

comparacion_global["Variación absoluta"] = (
    comparacion_global["Predicho 2026"]
    - comparacion_global["Real 2025"]
)

comparacion_global["Variación (%)"] = (
    comparacion_global["Variación absoluta"]
    / comparacion_global["Real 2025"]
) * 100

comparacion_global.round(2)
```

```python
# ==========================================
# COMPARACIÓN POR TIPO DE CRIMEN
# Q1-Q3 2025 REAL VS Q1-Q3 2026 PREDICHO
# ==========================================

real_tipo_2025 = (
    df_modelo[
        (df_modelo["año"] == 2025) &
        (df_modelo["num_trimestre"].isin([1, 2, 3]))
    ]
    .groupby("tipo de crimen")["conteo"]
    .sum()
)

pred_tipo_2026 = (
    df_pred_2026
    .groupby("tipo de crimen")["conteo_predicho"]
    .sum()
)

comparacion_tipo = pd.DataFrame({
    "Real Q1-Q3 2025": real_tipo_2025,
    "Predicho Q1-Q3 2026": pred_tipo_2026
})

comparacion_tipo["Variación"] = (
    comparacion_tipo["Predicho Q1-Q3 2026"]
    - comparacion_tipo["Real Q1-Q3 2025"]
)

comparacion_tipo["Variación (%)"] = (
    comparacion_tipo["Variación"]
    / comparacion_tipo["Real Q1-Q3 2025"]
) * 100

comparacion_tipo = comparacion_tipo.sort_values(
    "Variación (%)",
    ascending=False
)

comparacion_tipo.round(2)
```

```python
# ==========================================
# BENCHMARK EVENTOS DE BAJA FRECUENCIA
# VALIDACIÓN SOBRE 2025
# ==========================================

categorias_raras = [
    "Secuestro",
    "Homicidios dolosos y asesinatos consumados",
    "Homicidios dolosos y asesinatos en grado tentativa"
]

# Seleccionamos únicamente las tres categorías
df_raras = df_modelo[
    df_modelo["tipo de crimen"].isin(categorias_raras)
].copy()

# Solo 2025, porque queremos validar contra datos reales
test_raras = df_raras[
    df_raras["año"] == 2025
].copy()

# Método 1: mismo trimestre del año anterior
test_raras["pred_lag4"] = test_raras["lag_4"]

# Método 2: media últimos cuatro trimestres
test_raras["pred_media4"] = (
    test_raras[
        ["lag_1", "lag_2", "lag_3", "lag_4"]
    ].mean(axis=1)
)

# Método 3: mediana últimos cuatro trimestres
test_raras["pred_mediana4"] = (
    test_raras[
        ["lag_1", "lag_2", "lag_3", "lag_4"]
    ].median(axis=1)
)

print(
    "Observaciones:",
    len(test_raras)
)

test_raras[
    [
        "municipio",
        "tipo de crimen",
        "trimestre",
        "conteo",
        "pred_lag4",
        "pred_media4",
        "pred_mediana4"
    ]
].head(20)
```

```python
# ==========================================
# COMPARACIÓN GLOBAL DE LOS 3 MÉTODOS
# CORREGIDA: EXCLUYE FILAS SIN HISTÓRICO SUFICIENTE
# ==========================================

metodos = {
    "Lag 4 estacional": "pred_lag4",
    "Media móvil 4": "pred_media4",
    "Mediana móvil 4": "pred_mediana4"
}

# Nos quedamos solo con filas donde los 3 métodos
# tienen una predicción válida
test_raras_validas = test_raras.dropna(
    subset=[
        "pred_lag4",
        "pred_media4",
        "pred_mediana4"
    ]
).copy()

print("Observaciones originales:", len(test_raras))
print("Observaciones válidas para comparar:", len(test_raras_validas))

resultados_raras = []

for nombre, columna in metodos.items():

    y_real_raras = test_raras_validas["conteo"]
    y_pred_raras = test_raras_validas[columna]

    mae = mean_absolute_error(
        y_real_raras,
        y_pred_raras
    )

    rmse = np.sqrt(
        mean_squared_error(
            y_real_raras,
            y_pred_raras
        )
    )

    wape = (
        np.abs(
            y_real_raras.values -
            y_pred_raras.values
        ).sum()
        /
        np.abs(y_real_raras.values).sum()
    ) * 100

    resultados_raras.append({
        "Método": nombre,
        "MAE": mae,
        "RMSE": rmse,
        "WAPE (%)": wape
    })

resultados_raras = pd.DataFrame(
    resultados_raras
).sort_values(
    "WAPE (%)"
)

resultados_raras.round(3)
```

```python
# ==========================================
# COMPARACIÓN POR CATEGORÍA
# ==========================================

resultados_raras_tipo = []

for tipo, datos in test_raras_validas.groupby(
    "tipo de crimen"
):

    for nombre, columna in metodos.items():

        y_real_tipo = datos["conteo"]
        y_pred_tipo = datos[columna]

        mae = mean_absolute_error(
            y_real_tipo,
            y_pred_tipo
        )

        rmse = np.sqrt(
            mean_squared_error(
                y_real_tipo,
                y_pred_tipo
            )
        )

        wape = (
            np.abs(
                y_real_tipo.values -
                y_pred_tipo.values
            ).sum()
            /
            np.abs(y_real_tipo.values).sum()
        ) * 100

        resultados_raras_tipo.append({
            "Tipo de crimen": tipo,
            "Método": nombre,
            "MAE": mae,
            "RMSE": rmse,
            "WAPE (%)": wape,
            "Casos reales": y_real_tipo.sum()
        })

resultados_raras_tipo = pd.DataFrame(
    resultados_raras_tipo
).sort_values(
    ["Tipo de crimen", "WAPE (%)"]
)

resultados_raras_tipo.round(3)
```

```python
# ==========================================
# PREDICCIÓN RECURSIVA 2026
# CATEGORÍAS DE MUY BAJA FRECUENCIA
# MEDIANA MÓVIL DE 4 TRIMESTRES
# ==========================================

categorias_raras = [
    "Secuestro",
    "Homicidios dolosos y asesinatos consumados",
    "Homicidios dolosos y asesinatos en grado tentativa"
]

# Histórico real hasta Q4 2025
df_hist_raras = df_modelo[
    (df_modelo["año"] <= 2025) &
    (df_modelo["tipo de crimen"].isin(categorias_raras))
].copy()

series_raras_2026 = (
    df_hist_raras[
        df_hist_raras["año"] == 2025
    ][["municipio", "tipo de crimen"]]
    .drop_duplicates()
)

predicciones_raras_2026 = []

for trimestre_num in [1, 2, 3]:

    trimestre = f"Q{trimestre_num}"

    nuevas_predicciones = []

    for _, serie in series_raras_2026.iterrows():

        municipio = serie["municipio"]
        tipo_crimen = serie["tipo de crimen"]

        historial = df_hist_raras[
            (df_hist_raras["municipio"] == municipio) &
            (df_hist_raras["tipo de crimen"] == tipo_crimen)
        ].sort_values(
            ["año", "num_trimestre"]
        )

        if len(historial) < 4:
            continue

        ultimos_4 = historial["conteo"].tail(4)

        prediccion = ultimos_4.median()

        nuevas_predicciones.append({
            "municipio": municipio,
            "tipo de crimen": tipo_crimen,
            "año": 2026,
            "trimestre": trimestre,
            "num_trimestre": trimestre_num,
            "conteo_predicho": prediccion,
            "modelo": "Mediana móvil 4",
            "nivel_confianza": "Alta incertidumbre"
        })

    nuevas_predicciones = pd.DataFrame(
        nuevas_predicciones
    )

    predicciones_raras_2026.append(
        nuevas_predicciones.copy()
    )

    # Incorporamos las predicciones al histórico
    nuevas_filas_hist = nuevas_predicciones[
        [
            "municipio",
            "tipo de crimen",
            "año",
            "trimestre",
            "num_trimestre",
            "conteo_predicho"
        ]
    ].rename(
        columns={
            "conteo_predicho": "conteo"
        }
    )

    df_hist_raras = pd.concat(
        [
            df_hist_raras,
            nuevas_filas_hist
        ],
        ignore_index=True
    )


df_pred_raras_2026 = pd.concat(
    predicciones_raras_2026,
    ignore_index=True
)

print(
    "Predicciones raras generadas:",
    len(df_pred_raras_2026)
)

df_pred_raras_2026.head(10)
```

```python
# ==========================================
# DATASET PREDICTIVO HÍBRIDO DEFINITIVO
# ==========================================

# Copia de las predicciones generales
df_pred_final_2026 = df_pred_2026.copy()

# Etiquetamos el modelo general
df_pred_final_2026["modelo"] = "Gradient Boosting"
df_pred_final_2026["nivel_confianza"] = "Estándar"

# Eliminamos las categorías raras
# que fueron predichas inicialmente por Gradient Boosting
df_pred_final_2026 = df_pred_final_2026[
    ~df_pred_final_2026[
        "tipo de crimen"
    ].isin(categorias_raras)
].copy()

# Unimos las predicciones específicas
df_pred_final_2026 = pd.concat(
    [
        df_pred_final_2026,
        df_pred_raras_2026
    ],
    ignore_index=True
)

# Ordenamos
df_pred_final_2026 = df_pred_final_2026.sort_values(
    [
        "municipio",
        "tipo de crimen",
        "num_trimestre"
    ]
).reset_index(drop=True)

print(
    "Total predicciones finales:",
    len(df_pred_final_2026)
)

print(
    "\nPredicciones por modelo:"
)

print(
    df_pred_final_2026["modelo"].value_counts()
)

print(
    "\nPredicciones por trimestre:"
)

print(
    df_pred_final_2026[
        "trimestre"
    ].value_counts().sort_index()
)
```

```python
# ==========================================
# COMPARACIÓN FINAL POR TIPO DE CRIMEN
# MODELO HÍBRIDO
# ==========================================

pred_tipo_2026_final = (
    df_pred_final_2026
    .groupby("tipo de crimen")["conteo_predicho"]
    .sum()
)

comparacion_tipo_final = pd.DataFrame({
    "Real Q1-Q3 2025": real_tipo_2025,
    "Predicho Q1-Q3 2026": pred_tipo_2026_final
})

comparacion_tipo_final["Variación"] = (
    comparacion_tipo_final["Predicho Q1-Q3 2026"]
    - comparacion_tipo_final["Real Q1-Q3 2025"]
)

comparacion_tipo_final["Variación (%)"] = (
    comparacion_tipo_final["Variación"]
    / comparacion_tipo_final["Real Q1-Q3 2025"]
) * 100

comparacion_tipo_final = comparacion_tipo_final.sort_values(
    "Variación (%)",
    ascending=False
)

comparacion_tipo_final.round(2)
```
